```python
from ultralytics import YOLO
```

```python
import os
import torch

# 目标文件夹路径
target_folder = 'D:/projects/yolov11/from_data_2024_12_purple_only_v2(70%)/val'#未增广val219
target_folder1 = 'D:/projects/yolov11/val'#增广


# 初始化一个空列表来存储 jpg 文件的绝对路径
jpg_files = []

# 遍历目标文件夹
for root, dirs, files in os.walk(target_folder1):
    for file in files:
        if file.endswith('.jpg'):
            # 获取文件的绝对路径
            file_path = os.path.join(root, file)
            # 将绝对路径添加到列表中
            jpg_files.append(file_path)

# 打印结果
print(len(jpg_files))
```

```python
model = YOLO("D:/projects/yolov11/runs/classify/train3/weights/best.pt")  # load a custom model
num = len(jpg_files)//100
green = 0
T_green = 0
purple = 0 
T_purple = 0
# results = model(jpg_files[2300],batch = 1,device ='cpu')
for i in range(0,num):
    results = model(jpg_files[100*i:100*(i+1)],batch = 1,device ='cuda')
    # print(len(jpg_files[100*i:100*(i+1)]))
    for r in results:
        if r.path.endswith('a.jpg'):
            green += 1
            if r.probs.top1 == 0:
                T_green += 1
        else:
            purple += 1
            if r.probs.top1 == 1:
                T_purple += 1
results = model(jpg_files[num*100:],batch = 1,device ='cuda')
for r in results:
    if r.path.endswith('a.jpg'):
        green += 1
        if r.probs.top1 == 0:
            T_green += 1
    else:
        purple += 1
        if r.probs.top1 == 1:
            T_purple += 1
print(green,T_green,purple,T_purple)
```

```python
import os
import random

# 指定文件夹路径
folder_path = 'D:/projects/yolov11/from_data_2024_12_purple_only_v2(70%)_balance/train/green'

# 获取文件夹下所有jpg文件
jpg_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]

# 计算需要删除的文件数量
num_to_delete = max(len(jpg_files) - 200, 0)

# 随机选择需要删除的文件
files_to_delete = random.sample(jpg_files, num_to_delete)

# 删除选中的文件
for file in files_to_delete:
    file_path = os.path.join(folder_path, file)
    os.remove(file_path)
    print(f"Deleted: {file_path}")

print(f"Remaining jpg files: {len(jpg_files) - num_to_delete}")
```
